# Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#required libraries to get an overview of dataset with Python
```

```python
def extract(filename):
    file = "/files/Macroeconometrics/Temperature/dat/" + filename  # identify location
    text = open(file, 'r')  #open text file
    
    table=[]  #create an empty table where relevant data will be inserted
    for line in text:
        column = line.split()   #split text file into words using the in-built split function
        if line[0] == '#':   #ignore comments
            continue
        elif len(column) == 0: #ignore empty lines
            continue
        else:
            year = int(column[0])  #define year variable
            temp_jan = column[1]  
            temp_feb = column[2]
            temp_mar = column[3]
            temp_apr = column[4]   #Creating variables for each month from the columns of .txt file
            temp_may = column[5]
            temp_jun = column[6]
            temp_jul = column[7]
            temp_aug = column[8]
            temp_sep = column[9]
            temp_oct = column[10]
            temp_nov = column[11]
            temp_dec= column[12]
            table.append([year, temp_jan, temp_feb, temp_mar, temp_apr, temp_may, temp_jun, temp_jul, temp_aug, temp_sep, temp_oct, temp_nov, temp_dec])  #insert variables
    text.close()
    return(table)
```

# Creating Pandas Dataframe for Temperature

*You will notice some columns will start with capital letters and others in lower case. I did this to differentiate between string values and float values. In the original dataset, most values were stored as strings since we imported a .txt file. For these values I named the columns with capital letters. However, after converting the datatype to float values I also turned the columns into lower case to differentiate the two.*

```python
x = extract("temp_data.txt")     #using extract function above on the .txt file where the data is stored line by line.
```

```python
data = pd.DataFrame(x, columns=['Year', 'January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])    
data
```

```python
df = data.drop(range(12), axis='index')    #Drops observations before 1876, as this is the year where other datasets for mortality will begin
```

```python
df['january'] = df['January'].astype(float)
df['february'] = df['February'].astype(float)        #Here we convert string columns into floats so that we can do further manipulations in the future. (Probably a better way to do this using a for loop)
df['march'] = df['March'].astype(float)
df['april'] = df['April'].astype(float)
df['may'] = df['May'].astype(float)
df['june'] = df['June'].astype(float)
df['july'] = df['July'].astype(float)
df['august'] = df['August'].astype(float)
df['september'] = df['September'].astype(float)
df['october'] = df['October'].astype(float)
df['november'] = df['November'].astype(float)
df['december'] = df['December'].astype(float)
```

```python
df = df.drop(columns=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])   #Drops old string columns
```

```python
df
```

```python
df['max_temp'] = df[['january', 'february', 'march', 'april', 'may', 'june', 'july', 'august', 'september', 'october', 'november', 'december']].max(axis=1)   #Create max temp column
```

```python
df['min_temp'] = df[['january', 'february', 'march', 'april', 'may', 'june', 'july', 'august', 'september', 'october', 'november', 'december']].min(axis = 1)    #Create min temp column
```

```python
df['median_temp'] = round(df[['january', 'february', 'march', 'april', 'may', 'june', 'july', 'august', 'september', 'october', 'november', 'december']].median(axis = 1), 2)    #Create median temp column
```
