```python
import pandas as pd
import matplotlib.pyplot as plt
#required libraries to get an overview of dataset with Python
```

```python
def extract(filename):
    #filename = input("Please indicate what datafile you would like to plot:") #indicate file (obsolete if we are calling this as a function)
    file = "/files/Macroeconometrics/Temperature/dat/" + filename  # identify location
    text = open(file, 'r')  #open text file
    
    table=[]  #create an empty table where relevant data will be inserted
    for line in text:
        column = line.split()   #split text file into words using the in-built split function
        if line[0] == '#':   #ignore comments
            continue
        elif len(column) == 0: #ignore empty lines
            continue
        else:
            year = int(column[0])  #define year variable
            temp_jan = column[1]  #define year variable
            temp_feb = column[2]
            temp_mar = column[3]
            temp_apr = column[4]
            temp_may = column[5]
            temp_jun = column[6]
            temp_jul = column[7]
            temp_aug = column[8]
            temp_sep = column[9]
            temp_oct = column[10]
            temp_nov = column[11]
            temp_dec= column[12]
            table.append([year, temp_jan, temp_feb, temp_mar, temp_apr, temp_may, temp_jun, temp_jul, temp_aug, temp_sep, temp_oct, temp_nov, temp_dec])  #insert variables
    text.close()
    return(table)
```

```python
x = extract("temp_data.txt")
#x
```
